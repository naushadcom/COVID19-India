# TimeSeries Data Into DataFrame

```python
# Function to fetch JSON data from a URL
def fetch_json(url):
    response = requests.get(url)
    response.raise_for_status()  # Check if the request was successful
    return response.json()

# URL of the JSON timeseries data
url_timeseries =  "https://data.covid19india.org/v4/min/timeseries.min.json"

# Fetch the JSON data
timeseries = fetch_json(url_timeseries)

# Process the data and convert it into a pandas DataFrame
rows = []
for region, region_data in timeseries.items():
    for date, date_data in region_data['dates'].items():
        row = {
            'Region': region,
            'Date': date,
            'Confirmed': date_data.get('delta', {}).get('confirmed', None),
            'Confirmed (7-day)': date_data.get('delta7', {}).get('confirmed', None),
            'Total Confirmed': date_data.get('total', {}).get('confirmed', None)
        }
        rows.append(row)

# Convert the list of rows into a pandas DataFrame
df = pd.DataFrame(rows)

# Display the first few rows of the DataFrame
print(df)
```

### May 3rd Week Covid Data

```python
import requests
import json
from datetime import datetime, timedelta
import pandas as pd

# Step 1: Download and parse the JSON data
url = "https://data.covid19india.org/v4/min/timeseries.min.json"
response = requests.get(url)
data = response.json()

# Step 2: Define the date range for the third week of May
start_date = datetime(2020, 5, 21)
end_date = datetime(2020, 5, 26)
date_range = [(start_date + timedelta(days=i)).strftime('%Y-%m-%d') for i in range((end_date - start_date).days + 1)]

# Step 3: Extract data for the required dates
# Process the data and convert it into a pandas DataFrame
rows = []
for region, region_data in data.items():
    for date in date_range:
        if date in region_data['dates']:
            date_data = region_data['dates'][date]
            row = {
                'Region': region,
                'Date': date,
                'Confirmed': date_data.get('delta', {}).get('confirmed', None),
                'Confirmed (7-day)_May': date_data.get('delta7', {}).get('confirmed', None),
                'Total Confirmed': date_data.get('total', {}).get('confirmed', None)
            }
            rows.append(row)
        else:
            print(f"No data for {date} in region {region}")

# Convert the list of rows into a pandas DataFrame
May_df = pd.DataFrame(rows)
print(May_df)
```

# August 3rd Week Data

```python
import requests
import json
from datetime import datetime, timedelta
import pandas as pd

# Step 1: Download and parse the JSON data
url = "https://data.covid19india.org/v4/min/timeseries.min.json"
response = requests.get(url)
data = response.json()

# Step 2: Define the date range for the third week of August
start_date = datetime(2020, 8, 14)
end_date = datetime(2020, 8, 21)
date_range = [(start_date + timedelta(days=i)).strftime('%Y-%m-%d') for i in range((end_date - start_date).days + 1)]

# Step 3: Extract data for the required dates
# Process the data and convert it into a pandas DataFrame
rows = []
for region, region_data in data.items():
    for date in date_range:
        if date in region_data['dates']:
            date_data = region_data['dates'][date]
            row = {
                'Region': region,
                'Date': date,
                'Confirmed': date_data.get('delta', {}).get('confirmed', None),
                'Confirmed (7-day)_Aug': date_data.get('delta7', {}).get('confirmed', None),
                'Total Confirmed': date_data.get('total', {}).get('confirmed', None)
            }
            rows.append(row)
        else:
            print(f"No data for {date} in region {region}")

# Convert the list of rows into a pandas DataFrame
Aug_df = pd.DataFrame(rows)
print(Aug_df)

```

### Total States Data

```python
# loopng through the data for the total states data
states_total = []

for i in states.keys():
    if i not in 'TT':
        state = i
        population = int(states[i]['meta']['population'])
        deceased = int(states[i]['total']['deceased'])
        recovered = int(states[i]['total']['recovered'])
        tested = int(states[i]['total']['tested'])
        vaccinated1 = int(states[i]['total']['vaccinated1'])
        vaccinated2 = int(states[i]['total']['vaccinated2'])

        dictionary_state_total = {'Region' : state , 'state' : dic_state_name[state] , 'Population' : population, 'Deceased' : deceased, 'Recovered' : recovered, 'Tested' : tested, 'Vaccinated_1' : vaccinated1, ' vaccinated_2' :vaccinated2}
        states_total.append(dictionary_state_total)
        # print(states_total)

# creating dataframe of states total

states_total_df = pd.DataFrame(states_total)
print(len(states_total_df))
states_total_df


print("All numeric Calculation is here\n")
Calculation_of_numeric_value = states_total_df.describe()
Calculation_of_numeric_value = states_total_df.describe()
Calculation_of_numeric_value
```

```python
# Comparision between Mean and Median
plt.plot(Calculation_of_numeric_value.loc['mean'])
plt.plot(Calculation_of_numeric_value.loc['50%'])
plt.legend(['mean','median'])
plt.show()
```

# MEREGE DATAFRAMES

```python
# Merging DataFrame
covid_data_raw = pd.merge(states_total_df , May_df  , on = ['Region'] , how = 'inner')
covid_data_raw_merge_may = pd.merge(covid_data_raw , May_df , on = ['Region'] , how = 'inner')
covid_data = pd.merge(covid_data_raw_merge_may , Aug_df , on = ['Region'] , how = 'inner')

covid_data.fillna(0,inplace = True)
covid_data
```

## DATA ANALYSIS ON MERGED DATA

What are the total number of death records across the country.

```python
Total_Death = covid_data['Deceased'].sum()

print(f"Total number of death across the country is {Total_Death}")
```

What is the Average of Recovered people

```python
Avg_of_recovering  = covid_data['Recovered'].mean()
print(f"Average of Recovered people is {Avg_of_recovering}")
```

```python
#specifying the columns in the main merged DataFrame
covid_data.columns
```

```python
# Calculation Requirement for the analysis

covid_data['Death_per_100_cases'] = covid_data['Deceased'] / covid_data['Total Confirmed'] * 100  #This gives the  Deaths of per 100 cases

covid_data['Recovered_per_100_cases'] = covid_data['Recovered'] / covid_data['Total Confirmed'] * 100  #This gives the  Recovery Rate of per 100 cases


covid_data['1_Week_Change'] =  covid_data['Total Confirmed_x'] - covid_data['Total Confirmed_x'] + covid_data['Confirmed (7-day)_May_y']  #This gives the  Deaths of per 100 cases

covid_data['%Increase'] = (covid_data['1_Week_Change'] / covid_data['Total Confirmed_x']) * 100

```

```python
#Grouped the Required Data for Ananlysis
grouped_covid_data = covid_data.groupby('state').agg({'Population':'sum' ,'Total Confirmed':'sum' ,'Recovered':'mean', 'Death_per_100_cases':'sum','Recovered_per_100_cases':'mean', '1_Week_Change':'sum','%Increase': 'mean'  }).reset_index()

grouped_covid_data.fillna(0,inplace = True) # Removing NaN value and fill  0 inplace

grouped_covid_data.head()
```

```python
#Total confirmed Cases of the Country
Total_Confirmed_Cases = grouped_covid_data['Total Confirmed'].sum()

print(f"India total confirmed Cases  {Total_Confirmed_Cases}")

```
